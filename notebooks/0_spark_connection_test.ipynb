# Giai đoạn 0: Kiểm tra kết nối PySpark đến Cassandra

Notebook này xác minh rằng môi trường PySpark của chúng ta có thể kết nối thành công đến cơ sở dữ liệu Cassandra nguồn bằng cách sử dụng `spark-cassandra-connector`.

```python
import os
# Cấu hình Spark để tải gói Datastax Cassandra connector
os.environ['PYSPARK_SUBMIT_ARGS'] = '--packages com.datastax.spark:spark-cassandra-connector_2.12:3.4.1 pyspark-shell'

from pyspark.sql import SparkSession

print("Đang khởi tạo Spark Session...")
spark = SparkSession.builder \
    .appName("ChatDataMigrationTest") \
    .config("spark.cassandra.connection.host", "cassandra-source") \
    .config("spark.cassandra.connection.port", "9042") \
    .getOrCreate()

print("Spark Session đã sẵn sàng!")
```

```python
# Tải bảng chat_table từ keyspace chat_system trong Cassandra
df = spark.read \
    .format("org.apache.spark.sql.cassandra") \
    .options(table="chat_table", keyspace="chat_system") \
    .load()

print(f"Thành công! Tổng số bản ghi tìm thấy: {df.count()}")
df.show(5)
```
